Using Python, load the titanic dataset, convert the gender column into a numeric field, create a column that gives total number of relatives on board and remove duplicates from the table. 

```python
# Import data into python environment
filename = "../datasets/titanic3.xls"
df = pd.read_excel(filename)
df.head()
```

```python
# Convert gender to numeric field
def sex_to_numeric(x):
    if x=='male':
        return 0
    if x=='female':
        return 1

df['sex_val'] = df['sex'].apply(sex_to_numeric)
df.where()
df.head()
```

```python
# Create a column that gives total number of relatives on board 
df['relatives'] = df['sibsp'] + df['parch'] # Add counts of Spouse, children, siblings, parents
df.head()
```

```python
# Remove duplicates from the table. 
dupe = df['ticket'].duplicated() # Creates list of True/False values
df[dupe]               # Show rows where duplicated is True
```
